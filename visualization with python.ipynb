```python
## example of barplot
x = ['M', 'F']
height = [df[df.Sex=='male'].shape[0], df[df.Sex=='female'].shape[0]]
plt.bar(x = x, height= height, color = 'g')
plt.title('Barplot')
plt.xlabel('Gender')
plt.ylabel('Count')
```

```python
x = ['M', 'F']
height = [df[df.Sex=='male'].shape[0], df[df.Sex=='female'].shape[0]]
plt.barh(y = x, width= height, color = 'g')
plt.title('Barplot')
plt.ylabel('Gender')
plt.xlabel('Count')
```
