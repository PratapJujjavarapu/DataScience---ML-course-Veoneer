# Set_xticks,set_xticklabels,set_title,legend for axis subplot

```python
axs =plt.subplot(1,1,1)
axs.plot(np.random.randn(1000).cumsum(),'b--',label= 'cumsum')
ticks = axs.set_xticks([0,250,500,750,1000])
labels = axs.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
axs.set_title('My first matplotlib plot')
plt.legend(loc='best')
plt.show()
```

```python
axs =plt.subplot(1,1,1)
axs.plot(np.random.randn(1000).cumsum(),'r--',label= 'cumsum')
props = {
'title': 'My first matplotlib plot',
'xlabel': 'Stages'
}
axs.set(**props)
plt.show()
```
